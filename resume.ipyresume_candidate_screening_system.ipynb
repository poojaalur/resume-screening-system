```python
# Resume / Candidate Screening System
# =========================================

# Import Libraries
import pandas as pd
import re

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

```

```python
 #Step 1: Sample Resume Data
# -----------------------------------------

data = {
    "Candidate": [
        "Alice",
        "Bob",
        "Charlie"
    ],

    "Resume": [
        "Python SQL Machine Learning Flask Data Analysis",
        "Java Spring Boot MySQL Backend Development",
        "Python Deep Learning NLP TensorFlow SQL"
    ]
}

df = pd.DataFrame(data)
```

```python
# Step 2: Job Description
# -----------------------------------------

job_description = """
Looking for a Python Developer with skills in
Python, SQL, Flask, Machine Learning, and Data Analysis
"""

# -----------------------------------------
# Step 3: Text Cleaning Function
# -----------------------------------------

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

df["Cleaned_Resume"] = df["Resume"].apply(clean_text)

cleaned_job_description = clean_text(job_description)
```

```python


# -----------------------------------------
# Step 3: Text Cleaning Function
# -----------------------------------------

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text

df["Cleaned_Resume"] = df["Resume"].apply(clean_text)

cleaned_job_description = clean_text(job_description)
```

```python
# Step 4: TF-IDF Vectorization
# -----------------------------------------

all_text = df["Cleaned_Resume"].tolist() + [cleaned_job_description]

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(all_text)
```

```python
# Step 5: Cosine Similarity
# -----------------------------------------

resume_vectors = tfidf_matrix[:-1]
job_vector = tfidf_matrix[-1]

similarity_scores = cosine_similarity(resume_vectors, job_vector)

df["Match_Score"] = similarity_scores * 100
```

```python
# Step 6: Skill Gap Identification
# -----------------------------------------

required_skills = [
    "python",
    "sql",
    "flask",
    "machine learning",
    "data analysis"
]

def find_missing_skills(resume):
    missing = []

    for skill in required_skills:
        if skill not in resume.lower():
            missing.append(skill)

    return ", ".join(missing)

df["Missing_Skills"] = df["Resume"].apply(find_missing_skills)

```

```python
# -----------------------------------------
# Step 7: Candidate Ranking
# -----------------------------------------

df = df.sort_values(by="Match_Score", ascending=False)

df["Rank"] = range(1, len(df) + 1)

# -----------------------------------------
# Step 8: Display Final Results
# -----------------------------------------

print("\n===== Resume Screening Results =====\n")

print(df[[
    "Candidate",
    "Match_Score",
    "Missing_Skills",
    "Rank"
]])
```

```python
# Step 9: Best Candidate


best_candidate = df.iloc[0]

print("\n===== Best Candidate =====\n")

print("Candidate Name:", best_candidate["Candidate"])
print("Match Score:", round(best_candidate["Match_Score"], 2))
print("Missing Skills:", best_candidate["Missing_Skills"])
print("Rank:", best_candidate["Rank"])
```
